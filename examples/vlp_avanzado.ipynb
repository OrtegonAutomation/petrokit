### ðŸ”¹ Curvas VLP

Las curvas **VLP (Vertical Lift Performance)** representan la **caÃ­da de presiÃ³n** que experimenta el fluido desde el fondo hasta la superficie, en funciÃ³n del caudal.
Se calculan a partir de balances de energÃ­a que consideran **pÃ©rdidas por fricciÃ³n, aceleraciÃ³n y gradiente gravitacional**.

### ðŸ”¹ CorrelaciÃ³n de Beggs & Brill (1973)

Desarrollada para flujo multifÃ¡sico en tuberÃ­as inclinadas.
Se basa en una clasificaciÃ³n de **regÃ­menes de flujo** (segregado, intermitente, distribuido) y proporciona la fracciÃ³n de lÃ­quido retenido (**holdup**) y la pÃ©rdida de presiÃ³n total:

$$
\Delta P = \Delta P_{f} + \Delta P_{a} + \Delta P_{g}
$$

Donde:

- $( \Delta P_f )$: pÃ©rdidas por fricciÃ³n
- $( \Delta P_a )$: pÃ©rdidas por aceleraciÃ³n
- $( \Delta P_g )$: componente gravitacional

### ðŸ”¹ CorrelaciÃ³n de Hagedorn & Brown (1965)

Modelo empÃ­rico derivado de experimentos de flujo vertical bifÃ¡sico.
Particularmente Ãºtil en **pozos verticales** y condiciones de flujo burbujeante.

Ambas correlaciones se aplican comÃºnmente para estimar la **presiÃ³n de fondo fluyente (pwf)** en funciÃ³n del caudal.
